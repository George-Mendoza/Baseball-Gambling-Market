```python
def event_sum(df, column, string):
    event = []
    for row in df.loc[:, (column)].astype(str):
        if row == string:
            e = 1
        else:
            e = 0
        event.append(e)
    return event
```

```python
def outs(df, innings_col):
    a = 3*(df[innings_col].str[:-2].astype(int) - 1)
    b = df[innings_col].str[-1].astype(int) + 1
    o = a + b
    return o
```
